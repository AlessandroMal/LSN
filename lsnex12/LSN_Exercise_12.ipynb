## Exercise 12.1

Testare la *deep neural network* riportata in consegna, il cui compito è riconoscere una cifra disegnata in un immagine da $28\times 28$ pixel, variando l'ottimizzatore usato; discutere i risultati e come questi cambiano.

Per prima cosa vengono allocati e trattati i dati di modo da prepararli per essere usati dalla rete neurale: vengono allocate 60000 immagini di training e 10000 di validazione, che sono poi convertiti in forma categorica per essere usati da una rete che utilizza la *categorical cross-entropy* come funzione di loss.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

img_rows, img_cols = 28, 28 # number of pixels 
num_classes = 10 # 10 digits

(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data() #load data from mnist

print('X_train shape:', X_train.shape, end=';\t')
print('Y_train shape:', Y_train.shape)
print('X_test shape :', X_test.shape, end=';\t')
print('Y_test shape :', Y_test.shape)

X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols) #reshape data in order to use it at best
X_test = X_test.astype('float32') #single precision float
X_test /= 255 #normalize data
Y_test = keras.utils.to_categorical(Y_test, num_classes) #move to categorical data type (000...010...00)

X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols) #same for training data
X_train = X_train.astype('float32')
X_train /= 255
Y_train = keras.utils.to_categorical(Y_train, num_classes)
```

Viene strutturata la DNN traite le seguenti funzioni: si hanno quindi due layer *dense conncted* che portano a un layer di *dropout*, seguito da un ultimo layer connesso densamente con funzione di attivazione *softmax* (così ho ricondotto l'output a delle probabilità sull'ultimo layer)

```python
def createDNN():
    model=keras.models.Sequential()
    model.add(keras.layers.Dense(400, input_shape=(img_rows*img_cols,), activation='relu'))
    model.add(keras.layers.Dense(100, activation='relu'))
    model.add(keras.layers.Dropout(0.5))
    model.add(keras.layers.Dense(num_classes, activation='softmax'))
    return model

def compile_DNN(opt):
    mod=createDNN()
    mod.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    return mod
```

Il modello viene quindi compilato e allenato; di seguito sono riportati risultati e analisi per ogni ottimizzatore testato. Essendo il dataset molto ampio usiamo i mini-batches per guadagnare in costo computazionale, a scapito della precisione del calcolo dei gradienti; con tutti gli ottimizzatori fissiamo il numero di epoche uguale a $10$

### <font color='blue'> Adamax </font> 

```python
model_DNN=compile_DNN('Adamax')
batch_s, ep=32, 10
#train and save history
history=model_DNN.fit(x=X_train, y=Y_train, batch_size=batch_s, epochs=ep,
                      verbose=1, shuffle=True, validation_data=(X_test, Y_test))
```

Di seguito vengono mostrate le performance della rete con questo ottimizzatore, si vede che l'accuratezza è piuttosto buona: si attesta attorno al $98\%$ e la rete commette un singolo errore nel campione da 10 immagini mostrato di seguito

```python
score = model_DNN.evaluate(X_test, Y_test, verbose=1) # evaluate model
pred  = model_DNN.predict(X_test) #predictions of the model
# print performance
print()
print('Test loss:', round(score[0],5))
print('Test accuracy:', round(score[1],5))
```

## Exercise 12.2

Svluppo una rete neurale convoluzionale per eseguire il riconoscimento delle cifre scritta a mano: lo specchietto seguente riporta il codice utilizzato per definire la struttura della rete neurale.

Le **reti neurali convoluzionali** sono in grado di sfruttare alcune simmetrie nell'analisi dei sistemi, come un'inverianza per traslazioni, per cui possono essere particolarmente adatte in certi tipi di problemi come questi **problemi di classificazione**.

<img src='img/CNN.jpeg' width=850>

Nello specchietto di codice seguente preparo i dati da dare in pasto a questa nuova rete.

```python
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_sh = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_sh = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

Costruisco ora la struttura della rete neurale: ho scelto di utilizzare due layers convoluzionali, inframezzati da layers di pooling e dopo un *flattening* dei dati inserisco dei layers densi; nell'ultimo layers avrò di nuovo $10$ unità la cui accensione con una certa probabilità corrisponde alla probabilità di avere a che fare con una certa cifra.

```python
def createCNN():
    model=keras.models.Sequential()
    model.add(keras.layers.Conv2D(10, kernel_size=(5,5), input_shape=input_sh, activation='relu'))
    model.add(keras.layers.MaxPooling2D())
    model.add(keras.layers.Conv2D(16, kernel_size=(5,5), activation='relu'))
    model.add(keras.layers.MaxPooling2D())
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(256, activation='relu'))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(num_classes, activation='softmax'))
    return model

def compile_CNN(opt):
    mod=createCNN()
    mod.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    return mod
```

Sono quindi pronto per allenare la rete, sempre con lo stesso dataset, $10$ epoche e *mini-batch*; come ottimizzatore viene usato lo stochastic gradient descent.

```python
model_CNN=compile_CNN('sgd')
batch_s, ep=32, 10
#train and save history
history=model_CNN.fit(x=X_train, y=Y_train, batch_size=batch_s, epochs=ep,
                      verbose=1, shuffle=True, validation_data=(X_test, Y_test))
```

Qui sotto è ripotortato un scorcio degli output convoluzionali della rete, che mostra come il dato iniziale viene trasportato e manipolato all'interno della rete.

```python
def display_activation(activations, col_size, row_size, layer_index): 
    activation = activations[layer_index]
    activation_index=0
    fig, ax = plt.subplots(row_size, col_size,  figsize=(row_size*3,col_size))
    for row in range(0,row_size):
        for col in range(0,col_size):
            ax[row][col].imshow(activation[0, :, :, activation_index], cmap='gray', aspect='auto')
            activation_index += 1

layer_outputs = [layer.output for layer in model_CNN.layers]
activation_model = keras.models.Model(inputs=model_CNN.input, outputs=layer_outputs)
```

```python
test_index = 3
activations = activation_model.predict(X_test[test_index].reshape(1,28,28,1))

plt.imshow(X_test[test_index][:,:,0], cmap='gray')
display_activation(activations, 4, 2, 0)
# col_size x row_size must be <= Number of filters for the convolutional layer
```

Dai valori di loss e accuracy ottenuti, la rete convoluzionale sembra fare meglio delle DNN, come del resto è lecito apettarsi in questo tipo di problema; l'accuratezza del modello si avvicina al $99\%$ e non quest'ultimo non commette errori nel campione sottostante.

```python
score = model_CNN.evaluate(X_test, Y_test, verbose=1) # evaluate model
pred  = model_CNN.predict(X_test) #predictions of the model
# print performance
print()
print('Test loss:', round(score[0],5))
print('Test accuracy:', round(score[1],5))
```

```python
import matplotlib.pyplot as plt

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)
plt.figure(figsize=(15, 6)) 
for i in range(10):
    r=np.random.randint(0,X_test.shape[0])
    ax = plt.subplot(2, 5, i + 1)    
    plt.imshow(X_test[r, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted: {}".format(np.argmax(Y_test[r]), np.argmax(pred[r])))    
    plt.axis('off')
    
plt.show()
```

Qui sotto sono riportati gli andamenti di *loss* e accuratezza: approcciando le 10 epoche, la fase di test sembra essere vicina ad un punto estremante.

```python
plt.figure(figsize=(16, 7))

plt.subplot(121) #summarize history for accuracy
plt.title('Accuratezza')
plt.plot(np.arange(1,len(history.history['accuracy'])+1), history.history['accuracy'], lw=2.5)
plt.plot(np.arange(1,len(history.history['val_accuracy'])+1), history.history['val_accuracy'], lw=2.5)
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()

plt.subplot(122) #summarize history for loss
plt.title('Loss function')
plt.plot(np.arange(1,len(history.history['loss'])+1), history.history['loss'], lw=2.5)
plt.plot(np.arange(1,len(history.history['val_loss'])+1), history.history['val_loss'], lw=2.5)
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()
```
