## Algoritmo Genetico
Implementare un algoritmo genetico per risolvere il problema del *Travelling Salesman Problem*: date le coordinate di N punti (città) in un certo spazio, trovare il percorso chiuso con lunghezza minima in tutto lo spazio dei percorsi possibili; un problema di questo tipo è riconosciuto essere *np-hard*. In questo esercizio si affronterà il problema per
- 32 città su una circonferenza di raggio $r$
- 32 città all'interno di un quadrato di lato $l$

Le posizioni delle città vengono generate in maniera casuale purchè rispettino le condizioni al contorno dettate dal problema, dopodichè viene fissata una città di partenza e generata una popolazione di percorsi chiusi iniziale in maniera casuale; questi percorsi sono nella pratica dei vettori di indici da $1$ a $N-1$ in cui ogni indice appare una volta sola, la città $0$ è fissata per tutti gli individui: avrò quindi individui composti dalle permutazioni del vettore $ \left [1, 2, 3, \dots, n, \dots N-1 \right]$.

L'algoritmo genetico deve lavorare in modo da far progredire nelle generazioni successive gli individui con un *fitness* migliore, mantenere le mutazioni vantaggiose e scartare quelle svantaggiose. Il fitness per questo problema corrisponde alla minimizzazione della funzione
$$ L(\vec{x}_0, \vec{x}_1, \dots, \vec{x}_N) = \sum_{i=1}^N |\vec{x}_{i+1} - \vec{x}_i|\,,$$

o della somma di una certa potenza delle distanze, data la monotonicità di questa
$$ L_p (\vec{x}_0, \vec{x}_1, \dots, \vec{x}_N) = \sum_{i=1}^N |\vec{x}_{i+1} - \vec{x}_i|^p\,,$$

Vediamo più in dettaglio gli operatori di **selezione** e **mutazione** che permettono il miglioramento della popolazione generazione dopo generazione.

### <font color='red'> Operatori di selezione</font>
La selezione dei migliori individui per passare da una generazione alla successiva viene fatta da due operatori successivi:

1) *eliminatore proporzionale a $L_2$*. Questo selettore sceglie un elemento che sarà eliminato con probabilità proporzionale alla funzione $L_2$ definita in precedenza. Ho fissato il *coefficiente di estizione* a $0.8$, il che vuol dire che il $20\%$ della popolazione iniziale sopravviverà a questa selezione.

2) *below-average selector*. Questo selettore prende gli elementi sopravvissuti e screma ulteriormente il pool mantenendo soltanto gli elementi sotto la media, senza però mai scendere sotto il $5\%$ della popolazione, in tal caso la selezione viene arrestata. Questo mi permette di mantenere più individui "buoni" se questi hanno fitness simili, o di eliminare molti individui ormai "non buoni" se c'è un piccolo gruppo molto migliore del resto della popolazione.

### <font color='red'> Mutazioni, crossover e ripopolazione</font>
Per ripopolare la nuova generazione, l'algoritmo sceglie in maniera casuale un individuo sopravvissuto, lo clona e lo inserisce nella nuova generazione. Questo clone iniziale aggiunto può andare poi incontro a una (e una sola) mutazione con una certa probabilità. Gli operatori di mutazione sono:

- *Pair permutation* ($P = 0.1$), che consiste nello scambio di due città scelte casualmente, ad eccezione della città zero che non deve essere modificata.

- *Block permutation* ($P = 0.1$), che consiste nello scambio di due sottopercorsi della stessa lunghezza (lunghi minimo $1$ e al più la metà di $N-1$) distinti scelti casualmente.

- *Shift sub-path* ($P = 0.1$), che consiste nello slittamento di un sottopercorso (lungo minimo $1$ e al più $N-2$, escludendo così la città zero e l'altro sottopercorso che sarà al minimo unitario) lungo il percorso totale, sempre ad eccezione della città zero.

- *Reverse sub-path* ($P = 0.1$), che consiste nell'inversione dell'ordine di un sottopercorso proprio.

Dopo ogni generazione e mutazione di un nuovo individuo, c'è una probabilità $P=0.5$ di avere un **crossover** tra due nuovi individui scelti casualmente: scelto in maniera casuale un punto di taglio dei percorsi, le parti terminali dei due individui vengono scambiate, apportando così una maggior ricombinazione dei nuovi individui.

L'algoritmo è a **popolazione costante**: la nuova generazione è ripopolata tanto da tornare allo stesso numero di individui della precedente.

### <font color='blue'> Città lungo una Circonferenza</font>
Applico l'algoritmo genetico al caso di $32$ città disposte lungo una circonferenza di raggio $r=1$. Utilizzo una popolazione formata da $500$ percorsi e costruisco $1000$ generazioni successive per trovare il percorso più breve.

Nel grafico sottostante sono riportate la lunghezza del percorso migliore, la media delle lunghezze e la media dei quadrati delle lunghezze della metà migliore della popolazione in funzione delle generazioni create dall'algoritmo. Si può osservare che col passare delle generazioni, il pool si popola sempre più di individui con fitness migliore e le curve convergono al risultato ottimale.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
x = np.linspace(0, 1000, 100)

av, av2, minL = np.loadtxt('data/circ/path_length.dat', unpack=True)
print("Lunghezza del percorso migliore:", minL[-1])
x=np.arange(1,len(av)+1)

plt.plot(x, minL, label='best path', lw=2)
plt.plot(x, av, label=r'best half $ \langle L\rangle $', lw=2)
plt.plot(x, av2, label=r'best half $ \langle \sum _i l_i ^2 \rangle $', lw=2)
plt.title("Path length")
plt.legend()
plt.xlabel('#generation')
plt.ylabel('$L$')
plt.grid(True)
```

Il grafico sottostante mostra il percorso migliore dell'ultima generazione costruita dall'algoritmo. In questo caso è facile osservare che l'algoritmo converge a un buon minimo che potrebbe sembrare anche quello globale, ma anche se non lo fosse, il minimo locale trovato sarebbe comunque soddisfacente: la circonferenza sembra ben percorsa in uno dei due sensi, cammino che corrisponde al minimo globale per questo problema.

```python
import numpy as np
import matplotlib.pyplot as plt

data=open('data/circ/opt_path.dat')
strcities=data.readline()
strpath=data.readline()
citiesxy=np.fromstring(strcities, sep=' ')
bestpath=np.fromstring(strpath, dtype=np.int32, sep=' ')

xpath = [citiesxy[0]]
ypath = [citiesxy[1]]

for i in bestpath:
    xpath.append(citiesxy[2*i])
    ypath.append(citiesxy[2*i+1])

xpath.append(citiesxy[0])
ypath.append(citiesxy[1])

plt.figure(figsize=(12, 12))
plt.title("Città lungo una circonferenza, $r=1$")
plt.scatter(citiesxy[2::2], citiesxy[3::2], s=40, c='black') #città
plt.scatter(citiesxy[0], citiesxy[1], s=80, marker='D', c='red')
plt.plot(xpath, ypath, lw=2.5, color='green') #percorso
plt.show()
```

### <font color='blue'> Città in un Quadrato</font>
Applico l'algoritmo genetico al caso di $32$ città disposte casualmente all'interno di un quadrato di lato $l=1$. Utilizzo una popolazione formata da $500$ percorsi e costruisco $1000$ generazioni successive per trovare il percorso più breve.

Nel grafico sottostante sono riportate la lunghezza del percorso migliore, la media delle lunghezze e la media dei quadrati delle lunghezze della metà migliore della popolazione in funzione delle generazioni create dall'algoritmo. Si può osservare che col passare delle generazioni, il pool si popola sempre più di individui con fitness migliore e le curve sembrano convergere per lo meno a un buon minimo locale.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
x = np.linspace(0, 1000, 100)

av, av2, minL = np.loadtxt('data/square/path_length.dat', unpack=True)
print("Lunghezza del percorso migliore:", minL[-1])
x=np.arange(1,len(av)+1)

plt.plot(x, minL, label='best path', lw=2)
plt.plot(x, av, label=r'best half $ \langle L\rangle $', lw=2)
plt.plot(x, av2, label=r'best half $ \langle \sum _i l_i ^2 \rangle $', lw=2)
plt.title("Path length")
plt.legend()
plt.xlabel('#generation')
plt.ylabel('$L$')
plt.grid(True)
```

Il grafico sottostante mostra il percorso migliore dell'ultima generazione costruita dall'algoritmo. Come si può osservare l'algoritmo minimizza il percorso eliminando gli incroci tra le città minimizzando in modo significativo la lunghezza del cammnino. A occhio riesco a osservare due criticità: il nodo in alto a sinistra, il cui scioglimento può forse far guadagnare qualcosa, e il mini-cluster da tre città a ridosso della città $1$ osservando il percorso in senso antiorario; forse anche lì si potrebbe guadagnare un po' di terreno. Provando infatti a ripetere la simualzione con $2000$ generazioni il nodo si scioglie e anche altri due collegamenti vengono ottimizzati: sembra che riesca a raggiungere il minimo globale.

```python
import numpy as np
import matplotlib.pyplot as plt

filename=['data/square/opt_path.dat', 'data/square/opt_path2k.dat']
plottitle=['Città in una piazza quadrata, $l=1$, $gen=1000$', 'Città in una piazza quadrata, $l=1$, $gen=2000$']
subp=[121,122]
plt.figure(figsize=(15, 7.5))

for pl in range(len(subp)):
    data=open(filename[pl])
    strcities=data.readline()
    strpath=data.readline()
    citiesxy=np.fromstring(strcities, sep=' ')
    bestpath=np.fromstring(strpath, dtype=np.int32, sep=' ')

    xpath = [citiesxy[0]]
    ypath = [citiesxy[1]]

    for i in bestpath:
        xpath.append(citiesxy[2*i])
        ypath.append(citiesxy[2*i+1])

    xpath.append(citiesxy[0])
    ypath.append(citiesxy[1])

    plt.subplot(subp[pl])
    plt.title(plottitle[pl])
    plt.scatter(citiesxy[2::2], citiesxy[3::2], s=40, c='black') #città
    plt.scatter(citiesxy[0], citiesxy[1], s=80, marker='D', c='red')
    plt.plot(xpath, ypath, lw=2.5, color='green') #percorso

plt.show()
av, av2, minL = np.loadtxt('data/square/path_length2k.dat', unpack=True)
print("Lunghezza del percorso migliore con 2000 generazioni:", minL[-1])
```
