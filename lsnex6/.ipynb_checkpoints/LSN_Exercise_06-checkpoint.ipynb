### <font color=red> Equilibrazione </font>
La prima fase della simulazione consiste nell'equilibrazione del sistema a partire da $T=\infty$, cioè il caso in cui gli spin non hanno correlazione e ad ognuno di essi viene assegnato *up* o *down* casualmente con uguale probabilità. Questa equilibrazione verrà eseguita per ogni temperatura compresa tra $T=0.5$ e $T=2$ e sono stati scelti 100 punti in questo intervallo per mostrare l'andamento dei parametri in funzione di $T$.

Per capire quanto impiega l'algoritmo a oscillare attorno al valore esatto di energia, osservo i due casi estremi:
è ben visibile che per $T=2$ l'algoritmo impiega pochi passi per equilibrarsi attorno al valore atteso (caso mostrato qui sotto), mentre $T=0.515$ è il caso più difficile da equilibrare, come ci si aspetterebbe. Quest'ultimo infatti se svolto con Metropolis mostre un *acceptance rate* di di circa $3\%$, a differenza di $T=2$ che mostra $ar\simeq 50\%$.
E' stato quindi scelto un sistema di equilibrazione con 3 ripartenze, 1 blocco e $5\cdot 10^4$ mosse per blocco, di modo da equilibrare anche il sistema più restio in maniera consistente.

I grafici di seguito mostrano l'andamento dell'ultima ripartenza di equilibrazione.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

nstep = np.linspace(0, 1000, 1000)
Um = np.loadtxt('data/results/full_Emetro.out', delimiter=' ', unpack=True) #metropolis
Ug = np.loadtxt('data/results/full_Egibbs.out', delimiter=' ', unpack=True) #metropolis
Uup = -0.46*nstep**0
Udown = -0.97*nstep**0

Umdown= [Um[i] for i in range(0,1000)]
Umup= [Um[i] for i in range(1001,2001)]
del Um
Ugdown= [Ug[i] for i in range(0,1000)]
Ugup= [Ug[i] for i in range(1001,2001)]
del Ug

plt.figure(figsize=(16, 7.5))
plt.subplot(121)
plt.plot(nstep, Umdown, label='Metropolis') #Metropolis
plt.plot(nstep, Ugdown, label='Gibbs', alpha=0.85) #Gibbs
plt.plot(nstep, Udown, label='Valore analitico', color='black', linewidth=2)
plt.title('Energia interna istantanea, $h=0$, $T=0.5$')
plt.xlabel('#step Montecarlo')
plt.ylabel('U/N')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(nstep, Umup, label='Metropolis') #Metropolis
plt.plot(nstep, Ugup, label='Gibbs', alpha=0.85) #Gibbs
plt.plot(nstep, Uup, label='Valore analitico', color='black', linewidth=2)
plt.title('Energia interna istantanea, $h=0$, $T=2$')
plt.xlabel('#step Montecarlo')
plt.ylabel('U/N')
plt.legend()
plt.grid(True)

plt.show()
```

Si vede anche che l'energia della simulazione è quantizzata con $\Delta E_{min}=0.08$ e che gli switch sono molti di meno nel caso di temperatura bassa

### <font color=red> Simulazione</font>
Dopo aver equilibrato posso quindi prendere la configurazione finale e usarla per cominciare la vera simulazione e il calcolo dei parametri che mi interessano.

Per ogni temperatura simulata ho utilizzato il **data blocking** per calcolare valor medio e incertezza finale per ogni parametro: sono stati utilizzati $20$ blocchi da $5\cdot 10^5$ mosse ciascuno per il calcolo di ogni parametro.

E' bene specificare che una mossa consiste nel prendere uno spin della catena in maniera casuale e provare a girarlo seocndo le regole dell'algoritmo che sto utilizzando.

### <font color=blue> Energia interna </font>
In ogni tentativo di *switch*, ricalcolo l'energia secondo l'equazione del modello di Ising (con Metropolis devo osservare se viene accettata la mossa, altrimenti la configurazione rimane la stessa e riconto l'energia)
    $$ U(N,T) = -J\sum_{i=1}^N s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) \,,$$
e determino i valori medi e le incertezze statistiche su ciascun blocco per ogni valore della temperatura considerata.
La risoluzione analitica porta a
$$
\frac{U(N,T)}{N} = -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

Sono riportati qui di seguito i risultati per l'energia interna per spin ottenuta tramite gli algoritmi di Metropolis e di Gibbs in assenza di campo esterno $h=0$. L'accordo tra risultati attesi e simulazione sembra essere ottimale, le incertezze risultano poco visibili perchè molto piccole, evidentemente ho un alto numero di blocchi e per fortuna la convergenza al risultato esatto è ben verificata qui.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#calcolo analitico
points=100
T = np.linspace(0.4,2.1,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

#dati simulazioni
plt.figure(figsize=(10, 7.5))
temp, em, emerr = np.loadtxt('data/results/E_metro.out', delimiter=' ', unpack=True) #metropolis
temp, eg, egerr = np.loadtxt('data/results/E_gibbs.out', delimiter=' ', unpack=True) #Gibbs

plt.plot(T, e, color='green', label='risultato analitico')#, linewidth)
plt.errorbar(temp, em, yerr=emerr, label='Metropolis')#, linewidth)  #metropolis
plt.errorbar(temp, eg, yerr=egerr, label='Gibbs')#, linewidth) #Gibbs

plt.title('Energia interna, $h=0$')
plt.xlabel('T')
plt.ylabel('<H>/N')
plt.legend()
plt.grid(True)
plt.show()
```

### <font color=blue> Capacità termica </font>
In ogni step Montecarlo calcolo la capacità termica istantanea del sistema tramite la seguente equazione
    $$ C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2) \,,$$ 

e determino i valori medi e le incertezze statistiche su ciascun blocco per ogni valore della temperatura considerata.
La risoluzione analitica porta a
$$
\frac{C(N,T)}{N} =
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

Sono riportati qui di seguito i risultati per la capacità termica per spin ottenuta tramite gli algoritmi di Metropolis e di Gibbs in assenza di campo esterno $h=0$. La simulazione in questo caso sembra convergere a valori un po' differenti da quelli attesi, le incertezze risultano poco visibili perchè molto piccole, evidentemente ho un alto numero di blocchi e per fortuna la convergenza al risultato esatto è ben verificata qui.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#calcolo analitico
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

#dati simulazioni
plt.figure(figsize=(10, 7.5))
temp, Cm, Cmerr = np.loadtxt('data/results/C_metro.out', delimiter=' ', unpack=True) #Metropolis
temp, Cg, Cgerr = np.loadtxt('data/results/C_gibbs.out', delimiter=' ', unpack=True) #Gibbs

plt.plot(T, heat, color='green', label='risultato analitico')
plt.errorbar(temp, Cm, yerr=Cmerr, label='Metropolis') #Metropolis
plt.errorbar(temp, Cg, yerr=Cgerr, label='Gibbs')  #Gibbs

plt.title('Capacità termica, $h=0$')
plt.xlabel('T')
plt.ylabel('C/N')
plt.legend()
plt.grid(True)
plt.show()
```

### <font color=blue> Suscettività Magnetica </font>
In ogni step Montecarlo calcolo la suscettività magnetica istantanea del sistema tramite la seguente equazione
    $$ \chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0} \,,$$
(questo perchè $\left\langle \sum_{i=1,N}s_i \right\rangle^2 =0$ se il campo esterno è nullo) e determino i valori medi e le incertezze statistiche su ciascun blocco per ogni valore della temperatura considerata.
La risoluzione analitica porta a
$$
\chi(N,T) \underset{h=0}= \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

Sono riportati qui di seguito i risultati per la capacità termica per spin ottenuta tramite gli algoritmi di Metropolis e di Gibbs in assenza di campo esterno $h=0$. L'accordo tra risultati attesi e simulazione sembra essere ottimale, le incertezze risultano poco visibili perchè molto piccole, evidentemente ho un alto numero di blocchi e per fortuna la convergenza al risultato esatto è ben verificata qui.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#calcolo analitico
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

#dati simulazioni
plt.figure(figsize=(10, 7.5))
temp, chim, chimerr = np.loadtxt('data/results/X_metro.out', delimiter=' ', unpack=True) #Metropolis
temp, chig, chigerr = np.loadtxt('data/results/X_gibbs.out', delimiter=' ', unpack=True) #Gibbs

plt.plot(T, X, color='green', label='risultato analitico')
plt.errorbar(temp, chim, yerr=chimerr, label='Metropolis') #Metropolis
plt.errorbar(temp, chig, yerr=chigerr, label='Gibbs') #Gibbs

plt.title('Suscettività, $h=0$')
plt.xlabel('T')
plt.ylabel('$<\chi>$')
plt.legend()
plt.grid(True)
plt.show()
```

### <font color=blue> Magnetizzazione</font>
In ogni step Montecarlo calcolo la magnetizzazione istantanea del sistema tramite la seguente equazione
    $$ M(N,T,h)= \langle \sum_{i=1,N}s_i \rangle_{h} \,,$$
e determino i valori medi e le incertezze statistiche su ciascun blocco per ogni valore della temperatura considerata.
La risoluzione analitica porta a
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

dove i due autovalori della risoluzione analitica sono $\lambda_{1,2} = e^{\beta J} \cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$ e $Z= \lambda_1^N + \lambda_2^N$ è la funzione di partizione

Sono riportati qui di seguito i risultati per la capacità termica per spin ottenuta tramite gli algoritmi di Metropolis e di Gibbs in assenza di campo esterno $h=0$. L'accordo tra risultati attesi e simulazione sembra essere ottimale, le incertezze risultano poco visibili perchè molto piccole, evidentemente ho un alto numero di blocchi e per fortuna la convergenza al risultato esatto è ben verificata qui.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#calcolo analitico
h=0.02 #campo esterno
#autovalori:
l1 = np.exp(beta*J)*np.cosh(beta*h)+np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
l2 = np.exp(beta*J)*np.cosh(beta*h)-np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
Z = l1**Ns + l2**Ns
M = (np.exp(beta*J)*np.sinh(beta*h)*((l1**(Ns-1))*(1+np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))) 
        + (l2**(Ns-1))*(1-np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J)))))/(Z)

#dati simulazioni
plt.figure(figsize=(10, 7.5))
temp, Mm, Mmerr = np.loadtxt('data/results/M_metro.out', delimiter=' ', unpack=True) #Metropolis
temp, Mg, Mgerr = np.loadtxt('data/results/M_gibbs.out', delimiter=' ', unpack=True) #Gibbs

plt.plot(T, M, color='green', label='risultato analitico')
plt.errorbar(temp, Mm, yerr=Mmerr, label='Metropolis') #Metropolis
plt.errorbar(temp, Mg, yerr=Mgerr, label='Gibbs') #Gibbs

plt.title('Magnetizzazione $h=0,02$')
plt.xlabel('T')
plt.ylabel('<M>/N')
plt.legend()
plt.grid(True)

plt.show()
```

I parametri simulati hanno in generale un ottimo accordo, il quale è evidentemente più consistente e meno fluttuante a temperature più alte; il parametro meno accurato sembra essere la capacità termica. Le maggiori fluttuazioni a bassa temperatura sono ben visibili ad esempio se calcolo la magnetizzazione con $h=0$:

```python
h=0 #campo esterno
l1 = np.exp(beta*J)*np.cosh(beta*h)+np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
l2 = np.exp(beta*J)*np.cosh(beta*h)-np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
Z = l1**Ns + l2**Ns
M = (np.exp(beta*J)*np.sinh(beta*h)*((l1**(Ns-1))*(1+np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))) 
        + (l2**(Ns-1))*(1-np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J)))))/(Z)

#dati simulazioni
plt.figure(figsize=(10, 7.5))
temp, Mm, Mmerr = np.loadtxt('data/results/M0_metro.out', delimiter=' ', unpack=True) #Metropolis
temp, Mg, Mgerr = np.loadtxt('data/results/M0_gibbs.out', delimiter=' ', unpack=True) #Gibbs

plt.plot(T, M, color='green', label='risultato analitico')
plt.errorbar(temp, Mm, yerr=Mmerr, label='Metropolis') #Metropolis
plt.errorbar(temp, Mg, yerr=Mgerr, label='Gibbs') #Gibbs

plt.title('Magnetizzazione $h=0$')
plt.xlabel('T')
plt.ylabel('<M>/N')
plt.legend()
plt.grid(True)

plt.show()
```
