## Exercise 02.1
Calcolare il seguente integrale utilizzando i metodi Monte Carlo di con campionamento uniforme e *importance sampling*:
    $$ I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1 \,.$$

### <font color='blue'> Campionamento uniforme</font>
Posso calcolare $I$ con diverse distribuzioni di probabilità: iniziando dal caso più semplice, $p(x)=1/\Omega$ uniforme sul dominio di integrazione, posso stimare l'integrale come:
    $$ I = <\frac{\pi}{2}\cos ( \pi x/2 )>_{unif} \simeq \frac{1}{M} \sum_{i=1}^{M} \frac{\pi}{2}\cos ( \pi x_i/2 )    \,.$$
Il grafico qui sotto mostra la stima dell'integrale ottenuta utilizzando il metodo delle *medie a blocchi* per determinare media e deviazione standard in funzione del numero di valori casuali generati nell'intervallo.

Nella simulaizone sono stati utilizzati $10^5$ RN e suddivisi in $100$ blocchi.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
x,I,err= np.loadtxt('./data/Iuniform.dat', delimiter = ' ', unpack = 'True')
y = 1*x**0

plt.errorbar(x,I, yerr=err, label='integrazione numerica')  #grafico dell'integrale con barre di errore
plt.plot(x, y, color='green', label='risultato esatto')  #grafico del risultato esatto
plt.title('Campionamento uniforme')
plt.xlabel('#throws in [0,1)')
plt.ylabel('$<I>$')
plt.legend(loc=4)
plt.grid(True)
```

Posso confrontare la velocità di convergenza al valore esatto di questo metodo con quello dell'importance sampling.

### <font color='blue'> Importance sampling</font>
Nella tecnica dell'importance sampling si cerca di migliorare l'errore e quindi la velocità di covergenza dell'integrale Monte Carlo utilizzando una distribuzione di probabilità $p(x)$più simile alla funzione integranda da campionare: se la nuova cumulativa è invertibile analiticamente, posso applicare il metodo dell'**inversa della funzione cumulativa** per generare valori casuali distribuiti secondo la $p(x)$ che voglio utilizzare, a partire da una variabile casuale $y$ uniforme.
Nel calcolo seguente è stata scelta
    $$ p(x) = \lambda(1-x)  \,.$$

La cui cumulativa è invertibile molto facilmente: dopo aver imposto la normalizzazione di $p(x)$ su $[0,1]$ che mi da $\lambda =2 $ ottengo
      $$ x = 1 - \sqrt{1-y}  \,,$$
 
Aggiornando il calcolo Monte Carlo di I al caso di distribuzione non uniforme ottengo:
$$ I =<\frac{\pi}{2} \frac {\cos ( \pi x/2 )}{p(x)}>_{p(x)}\simeq \frac{1}{M} \sum_{i=1}^{M} \frac{\frac{\pi}{2}\cos ( \pi x_i/2 )}{p(x_i)}   \,.$$

Ancora una volta usiamo il **data blocking** per ottenere medie e errori, generando $10^5$ valori casuali e suddividendoli in $100$ blocchi.

Nel secondo grafico è possibile osservare che l'importance sampling riduce sensibilmente (ma può fare molto meglio se miglioro la $p(x)$!) le incertezze e il tempo per convergere a un risultato con una data incertezza

```python
plt.figure(figsize=(10, 7))
x,I,erris= np.loadtxt('./data/IIS.dat', delimiter =' ', unpack = 'True')
y = 1*x**0

plt.errorbar(x, I, yerr=erris, label='integrazione numerica')  #grafico dell'integrale con barre di errore
plt.plot(x, y, color='green', label='risultato esatto')  #grafico del risultato esatto
plt.title('Importance sampling')
plt.xlabel('#throws in [0,1)')
plt.ylabel('$<I>$')
plt.legend(loc=4)
plt.grid(True)
```

```python
plt.figure(figsize=(10, 7))

plt.plot(x, err, label='uniform sampling')  #incrtezze uniform sampling
plt.plot(x, erris, label='importance sampling')  #incertezze importance sampling

plt.title('Confronto errori statistici')
plt.xlabel('#throws in [0,1)')
plt.ylabel('rms')
plt.legend()
plt.grid(True)
print("dopo ", int(x[len(x)-1])," pseudo casuali:")
print("rms uniform= ",err[len(err)-1])
print("rms IS=      ",erris[len(erris)-1])
```

## Exercise 02.2
Simulare un numero elevato di random walk sia in un reticolo 3D con passo reticolare fissato $a$ sia nel continuo, osservare come varia il paramentro $\sqrt{\langle r^2 \rangle_N}$ in funzione del numero $N$ di passi.
### <font color='blue'> Random walk su reticolo cubico</font>
Nalla simulazione abbiamo fissato pe semplicità $a=1$ e ad ogni nuovo passo due numeri pseudocasuali scelgono direzione e verso di incremento, al quale quindi verrà aggiunto $a$. Ovviamente sarebbe possibile fare il tutto generando un solo RN per passo, ma per semplicità di codice è stata fatta questa scelta:

- Genero un valore casuale uniforme nell'intervallo $(0,3)$ e ne prendo la parte intera: a ogni intero associo una direzione.
- Genero un altro valore casuale uniforme nell'intervallo $(0,2)$ e associo:

se $r<1 \longrightarrow$  verso **positivo**

se $r>1 \longrightarrow$  verso **negativo**

Sono stati effettuati $10^4$ random walk differenti da $100$ step totali ciascuno, considerandoli come esperimenti diversi e indipendenti ho ottenuto una stima dell'estensione media del cammino $\sqrt{\langle r^2 \rangle_N}$ e, facendo la media del quadrato di $r_N ^2$, le relativa incertezza statistica in funzione del numero $N$ di passi effettuati.

Osservando i grafici seguenti, si nota un ottimo accordo con l'andamento atteso $f(N)=a\sqrt{N}$, inoltre è possibile effettuare  un fit dei dati per daterminare il paramentro $\alpha$ in $f(N)=\alpha \sqrt{N}$ e osservare se si accorda bene con il valore atteso $\alpha=1$: in effetti il parametro incognito dai dati risulta $\alpha=1.00323$, per cui il tipico comportamento diffusivo che si dimostra governare anche i cammini aleatori, è effettivamente verificato per i Random Walk simulati.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.figure(figsize=(10, 7))
rw, err= np.loadtxt('./data/RWlattice.dat', delimiter = ' ', unpack = 'True')

x=np.linspace(0,100,100)

#funzione con cui fittare l'andamento
def fitfunc(x, alpha):
    return alpha*(x**0.5)

popt, pcov = curve_fit(fitfunc, x, rw)  #fit del parametro alpha per il random walk nel reticolo
print(popt)

plt.errorbar(x, rw, err, label='$\sqrt{<r^2>}$', linewidth=2)  #varianza
plt.plot(x, fitfunc(x, *popt), color='red', label='fit', linewidth=2.5)  #fit della funzione 

plt.title("Reticolo cubico")
plt.xlabel('#step')
plt.ylabel('$\sqrt{<r^2>}$')
plt.grid(True)
plt.legend()
plt.show()
```

### <font color='blue'> Spazio continuo</font>
Fissiamo come nel caso precedente $|a|=1$, ma questa volta $\vec {a}$ è un qualsiasi vettore che punta in direzione casuale nello spazio tridimensionale, con modulo fissato. Sarà quindi necessario campionare in modo uniforme l'angolo solido attraverso la generazione dei valori casuali $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$ corrispondenti rispettivamente agli angoli **polare** e **azimutale** delle coordinate sferiche (vedi corollario seguente per la copertura uniforme della sfera unitaria).

Le direzioni non sono più quantizzate, quindi il cammino nelle 3 dimensioni è dato da
- $x_{n}= x_{n-1} + a \sin(\theta) \cos(\phi)$
- $y_{n}= y_{n-1} + a \sin(\theta) \sin(\phi)$
- $z_{n}= z_{n-1} + a \cos(\theta) $

con $x_n, y_n, z_n$ coordinate cartesiane della posizione dell'oggetto dopo $n$ passi.

Sono stati effettuati $10^4$ random walk differenti da $100$ step totali ciascuno, considerandoli come esperimenti diversi e indipendenti ho ottenuto una stima dell'estensione media del cammino $\sqrt{\langle r^2 \rangle_N}$ e, facendo la media del quadrato di $r_N ^2$, le relativa incertezza statistica in funzione del numero $N$ di passi effettuati.

Osservando i grafici seguenti, si nota un ottimo accordo con l'andamento atteso $f(N)=a\sqrt{N}$, inoltre è possibile effettuare  un fit dei dati per daterminare il paramentro $\alpha$ in $f(N)=\alpha \sqrt{N}$ e osservare se si accorda bene con il valore atteso $\alpha=1$: in effetti il parametro incognito dai dati risulta $\alpha=1.00752$, per cui il tipico comportamento diffusivo che si dimostra governare anche i cammini aleatori, è effettivamente verificato per i Random Walk simulati.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.figure(figsize=(10, 7))
rw, err= np.loadtxt('./data/RW3Dcont.dat', delimiter = ' ', unpack = 'True')

x=np.linspace(0,100,100)

#funzione con cui fittare l'andamento
def fitfunc(x, alpha):
    return alpha*(x**0.5)

popt, pcov = curve_fit(fitfunc, x, rw)  #fit del parametro alpha per il random walk nel reticolo
print(popt)

#grafico della varianza del rw e del fit
plt.errorbar(x, rw, err, label='$\sqrt{<r^2>}$', linewidth=2)  #varianza
plt.plot(x, fitfunc(x, *popt), color='red', label='fit', linewidth=2.5)  #fit della funzione 

plt.title("Spazio continuo")
plt.xlabel('#step')
plt.ylabel('$\sqrt{<r^2>}$')
plt.grid(True)
plt.legend()
plt.show()
```

### <font color='green'> Corollario: copertura uniforme della sfera unitaria</font>

Per poter campionare in modo uniforme l'angolo solido $\Omega$ devo determinare le densità di probabilità con cui generare gli angoli $\theta$ e $\phi$. Ho quindi una pdf congiunta che va normalizzata: ricordando la relazione tra i differenziali $d\Omega = \sin(\theta)d\theta d\phi$ ottengo la seguente equazione
    $$ P(\Omega) d\Omega = \frac{1}{4\pi} d\Omega = P(\theta, \phi) d\theta d\phi \,,$$
    
allora vedo immediatamente
    $$ P(\theta, \phi) = \frac{\sin(\theta)}{4\pi} \,.$$

Ora integrando separatamente sulle due variabili (quindi marginalizzando la pdf congiunta) ottengo le distribuzioni che sto cercando:
    $$p(\theta) = \int_0^{2\pi} P(\theta, \phi) d\phi  = \frac{sin(\theta)}{2} \,,$$ 
    $$p(\phi) = \int_0^{\pi}P(\theta, \phi) d\theta = \frac{1}{2\pi} \,.$$

funizoni per le quali è abbastanza semplice invertire la cumulativa e ottenere le due variabili casuali in funzione di un'altra variabile casuale, ma distribuita uniformemente su $[0,1]$:
$$ \theta = \arccos(2x-1)$$
$$ \phi = 2\pi y  \,.$$

Qui di seguito ho mostrato come l'ingannevole campionamento uniforme degli angoli porti a una copertura della sfera non uniforme, ma più concentrata attorno ai poli, mentre col campionamento sopra descritto si riesce a osservare una buona uniformità della copertura.

```python
from mpl_toolkits.mplot3d import Axes3D

M=2500
fig = plt.figure(figsize=(20, 20))
#campionamento sbagliato(uniforme) di theta e phi
theta=np.random.uniform(0.0, math.pi, M)
phi=np.random.uniform(0.0, 2*math.pi, M)
X=np.sin(theta)*np.cos(phi)
Y=np.sin(theta)*np.sin(phi)
Z=np.cos(theta)

ax = fig.add_subplot(2,2,1, projection='3d')
ax.scatter(X, Y, Z, marker='.')
ax.set_title("Side view")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, 30) #vista parallela all'equatore
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
ax.axes.zaxis.set_ticks([])

ax = fig.add_subplot(2,2,2, projection='3d')
ax.scatter(X, Y, Z, marker='.')
ax.set_title("Top view")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(90, 0) #vista parallela all'asse della sfera
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
ax.axes.zaxis.set_ticks([])
plt.show()
```

Evidente maggior concentrazione intorno ai poli della sfera!

Invece col campionamento descritto in corollario:

```python
fig = plt.figure(figsize=(20, 20))
#campionamento uniforme corretto dell'angolo solido
x=np.random.uniform(0.0, 1.0, M)
theta=np.arccos(2*x-1)
phi=np.random.uniform(0.0, 2*math.pi, M)
X=np.sin(theta)*np.cos(phi)
Y=np.sin(theta)*np.sin(phi)
Z=np.cos(theta)

ax = fig.add_subplot(2,2,3, projection='3d')
ax.scatter(X, Y, Z, c='tab:green', marker='.')
ax.set_title("Side view")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
ax.axes.zaxis.set_ticks([])
ax.view_init(0,30) #vista parallela all'equatoriale

ax = fig.add_subplot(2,2,4, projection='3d')
ax.scatter(X, Y, Z, c='tab:green', marker='.')
ax.set_title("Top view")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
ax.axes.zaxis.set_ticks([])
ax.view_init(90, 0) #vista parallela all'asse della sfera
plt.show()
```
